```python
import pandas as pd

column_names = [
    'id',
    'name',
    'formatted_address',
    'lat',
    'lng',
    'types',
    'rating',
    'user_rating_count',
    'price',
    'operational',
    'province'
]

df = pd.read_csv(
    'places_ankara.csv',
    header=None,
    names=column_names
)

df.head()
```

```python
df.info()
```

```python
types_dummies = df['types'].str.strip('[]').str.replace('"', '').str.get_dummies(sep=', ')

df = pd.concat([df, types_dummies], axis=1)

df = df.drop('types', axis=1)

df.head()
```

```python
popular_types_sorted = types_dummies.sum().sort_values(ascending=False)

pd.set_option('display.max_rows', None)

print(popular_types_sorted)

pd.reset_option('display.max_rows')
```
